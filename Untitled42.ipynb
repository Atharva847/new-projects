```python
'''
Q7. Can you explain the geometric interpretation of eigenvectors and eigenvalues?
Geometrically:

Eigenvalues represent the magnitude of stretching or compressing along the eigenvector's direction.
Eigenvectors indicate the directions that remain invariant (except for scaling) under the transformation defined by the matrix.
For example, in a 2D space, if an eigenvector points in the direction of a line, applying the matrix will stretch or compress points along that line, while points perpendicular to it will change direction and not remain invariant.
'''
```
